# Tensorflow Ex-5

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import logging
logging.getLogger("tensorflow").setLevel(logging.WARNING) ## Disable the warnings

import numpy as np
```

```python
n_f = 10
n_d_n = 3
```

```python
x = tf.placeholder(tf.float32,(None,n_f))
```

```python
b = tf.Variable(tf.zeros([n_d_n]))
```

```python
w = tf.Variable(tf.random_normal([n_f,n_d_n]))
# y = mx + c
xw = tf.matmul(x,w)
```

```python
z = tf.add(xw,b)
```

```python
#Activation Function
a = tf.sigmoid(z)
```

```python
# Variable init
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    merged = tf.summary.merge_all()
    writer = tf.summary.FileWriter("/tmp/tensorflowlogs",sess.graph)
    sess.run(init)
    layer_out = sess.run(a,feed_dict={x : np.random.random([1, n_f])})
```

```python
print(layer_out)
```
